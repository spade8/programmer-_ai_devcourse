- Ï£ºÏñ¥ÏßÑ ÎßÅÌÅ¨Î•º ÌÜµÌï¥ `csv` ÌååÏùºÏùÑ Îã§Ïö¥Î∞õÍ≥† `pd.read_csv` Î©îÏÜåÎìúÎ•º ÌôúÏö©ÌïòÏó¨ `csv` ÌååÏùºÏóê Ï†ëÍ∑ºÌïúÎã§.

```python
netflix_data = pd.read_csv("./netflix_titles.csv")
netflix_data
```

### Mandatory Part
[Îã§Ïùå Îç∞Ïù¥ÌÑ∞](https://www.kaggle.com/datasets/rishidamarla/bitcoin-prices-20112015) Îã§Ïùå Îç∞Ïù¥ÌÑ∞Í∞Ä Ï£ºÏñ¥Ï°åÏùÑ Îïå 2016.6 ~ 2017.6 Í∏∞Í∞ÑÏùò 5-MA(Moving Average) ÎπÑÌä∏ÏΩîÏù∏ Í∞ÄÍ≤© Í∑∏ÎûòÌîÑÎ•º Í∑∏Î†§Ï£ºÏÑ∏Ïöî.

 - ÏÑ†Ïùò ÏÉâÍπîÏùÄ **ÎÖ∏ÎûÄÏÉâ(#f2a900)** ÏúºÎ°ú Ìï¥ÏïºÌï©ÎãàÎã§.
 - x-axis labelÍ≥º y-axis labelÏùÄ Ï†ÅÏ†àÌïú ÌÖçÏä§Ìä∏Î°ú Ï∂îÍ∞ÄÎêòÏñ¥Ïïº Ìï©ÎãàÎã§.
 - Ïù¥ÎèôÌèâÍ∑†Ïùò Í∏∞Ï§ÄÏùÄ `Open` ColumnÏúºÎ°ú Í≥ÑÏÇ∞Ìï©ÎãàÎã§.
 - Ïù¥Ïô∏ÏóêÎèÑ Í∑∏ÎûòÌîÑÎ•º ÏïåÏïÑÎ≥¥Í∏∞ ÏâΩÍ≤å ÌïòÍ∏∞ ÏúÑÌïú Ï†ïÎ≥¥(Text, Facet Îì±)ÏùÑ Ï∂îÍ∞ÄÌïòÎäî Í≤ÉÎèÑ Ï¢ãÏäµÎãàÎã§.
 
üí° Ïù¥ÎèôÌèâÍ∑†(Moving Average)Î≤ïÏùÄ ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞Î•º ÌëúÌòÑÌïòÎäî Îç∞Ïóê ÎßéÏù¥ ÏÇ¨Ïö©ÌïòÎäî Î∞©Î≤ï Ï§ë ÌïòÎÇòÏûÖÎãàÎã§.

 - Î®ºÏ†Ä Î¨∏Ï†ú Ìï¥Í≤∞Ïóê ÌïÑÏöîÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨Î•º Ìò∏Ï∂úÌïúÎã§.

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

- Ï£ºÏñ¥ÏßÑ ÎßÅÌÅ¨Î•º ÌÜµÌï¥ `csv` ÌååÏùºÏùÑ Îã§Ïö¥Î∞õÍ≥† `pd.read_csv` Î©îÏÜåÎìúÎ•º ÌôúÏö©ÌïòÏó¨ `csv` ÌååÏùºÏóê Ï†ëÍ∑ºÌïúÎã§.
- Ï£ºÏñ¥ÏßÑ ÎÇ†Ïßú Ï°∞Í±¥Ïóê ÎßûÍ≤å DataFrameÏùÑ Ìé∏ÏßëÌïúÎã§.
- DataÎ•º ÎÇ†ÏßúÎ•º Í∏∞Ï§ÄÏúºÎ°ú Ï†ïÎ†¨ÌïúÌõÑ `index`Î•º Ïû¨ÏÑ§Ï†ïÌïúÎã§.

```python
bitcoin_data = pd.read_csv("./BitCoin.csv")
bitcoin_data = bitcoin_data[(bitcoin_data['Date']>='2016-06-01') & (bitcoin_data['Date']<'2017-07-01')]
bitcoin_data = bitcoin_data.sort_values('Date')
bitcoin_data.reset_index(inplace=True, drop=True)
bitcoin_data
```

- `.to_datetime` Ìï®ÏàòÎ•º Ïù¥Ïö©ÌïòÏó¨ `Date` columnÏùò Í∞íÏùÑ ÌòïÎ≥ÄÌôò Ìï¥Ï§ÄÎã§.

```python
bitcoin_data['Date'] = pd.to_datetime(bitcoin_data['Date'])
bitcoin_data
```

- Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶∞Îã§. 

```python
ma5 = bitcoin_data['Open'].rolling(window=5).mean()
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(15,8))
plt.plot(bitcoin_data['Date'], ma5, color="#f2a900")
plt.title(" MA(Moving Average) ÎπÑÌä∏ÏΩîÏù∏ Í∞ÄÍ≤© Í∑∏ÎûòÌîÑ")
plt.xlabel("Date")
plt.ylabel("Bitcoin Moving Average")
plt.grid(True, linestyle='--')
plt.show()
```

- Ï£ºÏñ¥ÏßÑ ÎßÅÌÅ¨Î•º ÌÜµÌï¥ `csv` ÌååÏùºÏùÑ Îã§Ïö¥Î∞õÍ≥† `pd.read_csv` Î©îÏÜåÎìúÎ•º ÌôúÏö©ÌïòÏó¨ `csv` ÌååÏùºÏóê Ï†ëÍ∑ºÌïúÎã§.
- Ï£ºÏñ¥ÏßÑ ÎÇ†Ïßú Ï°∞Í±¥Ïóê ÎßûÍ≤å DataFrameÏùÑ Ìé∏ÏßëÌïúÎã§.

- `.to_datetime` Ìï®ÏàòÎ•º Ïù¥Ïö©ÌïòÏó¨ `Date` columnÏùò Í∞íÏùÑ ÌòïÎ≥ÄÌôò Ìï¥Ï§ÄÎã§.

```python
eth_data = pd.read_csv("./ETH_day.csv")
eth_data = eth_data[(eth_data['Date']>='2016-06-01') & (eth_data['Date']<'2017-07-01')]
eth_data = eth_data.sort_values('Date')
eth_data.reset_index(inplace=True, drop=True)
eth_data['Date'] = pd.to_datetime(eth_data['Date'])
eth_data
```

- Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶∞Îã§. 

```python
ma5_bitcoin = bitcoin_data['Open'].rolling(window=5).mean()
ma5_eth = eth_data['Open'].rolling(window=5).mean()
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(15,8))
plt.plot(bitcoin_data['Date'], ma5_bitcoin, color="#f2a900")
plt.plot(eth_data['Date'], ma5_eth, color="#3c3c3d")
plt.title("MA(Moving Average) ÎπÑÌä∏ÏΩîÏù∏ & Ïù¥ÎçîÎ¶¨ÏõÄ Í∞ÄÍ≤© Í∑∏ÎûòÌîÑ")
plt.xlabel("Date")
plt.ylabel("Moving Average")
plt.grid(True, linestyle='--')
plt.legend(labels=['Bitcoin','Ethereum'])
plt.show()
```
